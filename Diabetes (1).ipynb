```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop(['class','Alopecia','Itching'],axis=1)
X=pd.get_dummies(data=X,drop_first='True')
y = df['class']
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=101)
```

```python
from sklearn.svm import SVC
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
svc = SVC(class_weight='balanced')
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C':[0.001,0.01,0.1,0.5,1],'gamma':['scale','auto']}
grid = GridSearchCV(svc,param_grid)
```

```python
grid.fit(scaled_X_train,y_train)
```

```python
grid.best_params_
```

```python
grid_pred = grid.predict(scaled_X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
confusion_matrix(y_test,grid_pred)
```

```python
#Classification Report For SVM Classification Algorithm
```

```python
print(classification_report(y_test,grid_pred))
```

```python
accuracy_score(y_test,grid_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
operations = [('scaler',scaler),('knn',knn)]
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(operations)
```

```python
k_values = list(range(1,20))
```

```python
param_grid = {'knn__n_neighbors': k_values}
```

```python
from sklearn.model_selection import GridSearchCV
full_cv_classifier = GridSearchCV(pipe,param_grid,cv=5,scoring='accuracy')
full_cv_classifier.fit(X_train,y_train)
```

```python
full_cv_classifier.best_estimator_.get_params()
```

```python
scaler = StandardScaler()
knn1 = KNeighborsClassifier(n_neighbors=1)
operations = [('scaler',scaler),('knn1',knn1)]
```

```python
pipe = Pipeline(operations)
```

```python
pipe.fit(X_train,y_train)
```

```python
pipe_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
#Classification Report For KNN Classification Algorithm
```

```python
print(classification_report(y_test,pipe_pred))
```

```python
accuracy_score(y_test,pipe_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
```
